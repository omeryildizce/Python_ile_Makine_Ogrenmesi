```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
veriler = pd.read_csv("../veri/Ads_CTR_Optimisation.csv")
veriler
```

```python
import math
N = 10_000
d = 10
oduller = [0] * d
toplam = 0
tiklamalar = [0] * d
secilenler = []
for n in range(0, N):
    ad = 0
    max_ucb = 0
    for i in range(d):
        if(tiklamalar[i]>0):
            ortalama = oduller[i] / tiklamalar[i]
            delta = math.sqrt(3/2* math.log(n) / tiklamalar[i]) 
            ucb = ortalama + delta
        else:
            ucb = N * d
        if max_ucb < ucb:
            max_ucb = ucb
            ad = i
    secilenler.append(ad)
    tiklamalar[ad] = tiklamalar[ad] + 1
    odul = veriler.values[n, ad]
    oduller[ad] = oduller[ad] + odul
    toplam += odul

print("Toplam Ödul: ")
print(toplam)
```

```python
plt.hist(secilenler)
plt.show()
```
