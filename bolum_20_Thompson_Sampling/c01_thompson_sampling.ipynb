```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
veriler = pd.read_csv("../veri/Ads_CTR_Optimisation.csv")
veriler
```

```python
import random
N = 10_000
d = 10 
toplam = 0
tiklamalar = [0] * d
secilenler = []
birler = [0] * d
sifirlar = [0] * d
for n in range(0, N):
    ad = 0
    max_th = 0
    for i in range(d):
        rasbeta = random.betavariate(birler[i] + 1, sifirlar[i] + 1)
        if rasbeta > max_th:
            max_th = rasbeta
            ad = i
            
    secilenler.append(ad)
    tiklamalar[ad] = tiklamalar[ad] + 1
    odul = veriler.values[n, ad]
    if odul == 1:
        birler[ad] = birler[ad] + 1
    else:
        sifirlar[ad] = sifirlar[ad] + 1 
    toplam += odul

print("Toplam Ödul: ")
print(toplam)
```

```python
plt.hist(secilenler)
plt.show()
```
